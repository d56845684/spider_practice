```python
def scrapy(dict_, apart_name, sep_name):
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    for i in soup.find_all('tr',{'class':'Grid_Row'}):
        dict_['計畫類別'].append(apart_name)
        dict_['學門類別'].append(sep_name)
        for index, line in enumerate(i.find_all('td')):
            if index == 0:
                dict_['計畫年度'].append(line.text)
            elif index == 1:
                dict_['主持人姓名'].append(line.text)
            elif index == 2:
                dict_['執行機關'].append(line.text)
            elif index == 3:
                for number in range(200):
                    if not line.find('span', {'id':f'wUctlAwardQueryPage_grdResult_lblAWARD_PLAN_CHI_DESCc_{number}'}):
                        continue
                    else:
                        pg_name = line.find('span', {'id':f'wUctlAwardQueryPage_grdResult_lblAWARD_PLAN_CHI_DESCc_{number}'}).text
                        plan_time = line.find('span', {'id':f'wUctlAwardQueryPage_grdResult_lblAWARD_ST_ENDc_{number}'}).text
                        total_money = line.find('span', {'id':f'wUctlAwardQueryPage_grdResult_lblAWARD_TOT_AUD_AMTc_{number}'}).text
                        dict_['計畫名稱'].append(pg_name)
                        dict_['執行起迄'].append(plan_time)
                        dict_['總核定金額'].append(total_money)
#                         print(f'{i}, {pg_name}, {plan_time}, {total_money}')
#                         dict_['內容'].append(line.text)
#     soup_2 = BeautifulSoup(driver.page_source, 'html.parser')
    for _ in soup.find_all('tr',{'class':'Grid_AlternatingRow'}):
        dict_['計畫類別'].append(apart_name)
        dict_['學門類別'].append(sep_name)
        for index_2, line_2 in enumerate(_.find_all('td')):
            if index_2 == 0:
                dict_['計畫年度'].append(line_2.text)
            elif index_2 == 1:
                dict_['主持人姓名'].append(line_2.text)
            elif index_2 == 2:
                dict_['執行機關'].append(line_2.text)
            elif index_2 == 3:
                for number_ in range(200):
                    if not line_2.find('span', {'id':f'wUctlAwardQueryPage_grdResult_lblAWARD_PLAN_CHI_DESCc_{number_}'}):
                        continue
                    else:
                        pg_name = line_2.find('span', {'id':f'wUctlAwardQueryPage_grdResult_lblAWARD_PLAN_CHI_DESCc_{number_}'}).text
                        plan_time = line_2.find('span', {'id':f'wUctlAwardQueryPage_grdResult_lblAWARD_ST_ENDc_{number_}'}).text
                        total_money = line_2.find('span', {'id':f'wUctlAwardQueryPage_grdResult_lblAWARD_TOT_AUD_AMTc_{number_}'}).text
                        dict_['計畫名稱'].append(pg_name)
                        dict_['執行起迄'].append(plan_time)
                        dict_['總核定金額'].append(total_money)
#                         print(f'{i}, {pg_name}, {plan_time}, {total_money}')
#                         dict_['內容'].append(line_2.text)
```
